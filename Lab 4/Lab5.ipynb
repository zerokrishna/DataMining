```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.randn()
print(f"{w1} , {w2} , {b}")
```

```python
# Dataset
x1 = 1.5
x2 = 1
y = 5
```

```python
epochs = 20
alpha = 0.1
errorList = []

for epoch in range(epochs):
    # forward pass
    y_pred = x1 * w1+ x2* w2 +b
    
    #Compute error
    error = (y-y_pred)**2
    
    # compute gradient 
    deltaW1 = -2*(y-y_pred)*x1
    deltaW2 = -2*(y-y_pred)*x2
    deltab = -2*(y-y_pred)
    # Backpropagate
    w1 = w1-(deltaW1 * alpha)
    w2 = w2-(deltaW2 * alpha)
    b = b-(b * alpha)
    
    
    # Append the error for plotting loss curve
    errorList.append(error)
    print(f"Epoch {epoch+1} | Loss {error:-4f}")
    
```

```python
y_pred = x1 * w1 + x2 * w2 +b
print(y_pred)
```

```python
# plotting the loss curve
plt.figure(figsize=(8,6))
plt.plot(range(epochs),errorList)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks(ticks = range(epochs))
plt.title("Epochs vs Loss")
plt.show()
```
