```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w1 = np.random.randn()
w2 = np.random.randn()
w3 = np.random.randn()
w4 = np.random.randn()
w5 = np.random.randn()
w6 = np.random.randn()
b1 = np.random.randn()
b2 = np.random.randn()
b3 = np.random.randn()
```

```python
# Dataset
x1 = 5.5
x2 = 6.2
y = 11.2
```

```python
epochs = 14
alpha = 0.1
errorList = []

for epoch in range(epochs):
    # Forward pass
    h1 = x1*w1 + x2 * w3 + b1
    h2 = x1*w2 + x2 * w4 + b2
    y_pred = h1 * w5 + h2 * w6 + b3
    
    # Error Computation
    error = (y - y_pred)**2
    delta = -2*(y-y_pred)
    # compute gradient
    deltaW0 = delta*x1*w5
    deltaW1 = delta*x1*w6
    deltaW2 = delta*x2*w5
    deltaW3 = delta*x2*w6
    deltaW4 = delta*h1
    deltaW5 = delta*h2
    deltaB0 = delta * w5
    deltaB1 = delta * w6
    deltaB2 = delta
    
    # Backpropagate  
    w1 = w1-(deltaW0 * alpha)
    w2 = w2-(deltaW0 * alpha)
    w3 = w3-(deltaW0 * alpha)
    w4 = w4-(deltaW0 * alpha)
    w5 = w5-(deltaW0 * alpha)
    w6 = w6-(deltaW0 * alpha)
    b1 = b1-(deltaW0 * alpha)
    b2 = b2-(deltaW0 * alpha)
    b3 = b3-(deltaW0 * alpha)
    
    # Append the error for plotting loss curve
    errorList.append(error)
    print(f"Epoch {epoch+1} | Loss {error:-4f}")
```

```python

y_pred = h1 * w5 + h2 * w6 + b3
print(y_pred)
```

```python
# plotting the loss curve
plt.figure(figsize=(8,6))
plt.plot(range(epochs),errorList)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks(ticks = range(epochs))
plt.title("Epochs vs Loss")
plt.show()
```
