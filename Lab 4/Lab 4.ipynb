```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Initialize weight
w = np.random.randn()
print(w)
```

```python
# Dataset
x = 1.5
y = 0.5
```

```python
epochs = 10
alpha = 0.1
errorList = []
for epoch in range(epochs):
    # Forward pass
    y_pred = x*w
    # Compute error
    error = (y-y_pred)**2
    # Compute Gradients
    deltaW = -2*(y-y_pred)*x
    # Backpropagate
    w = w-(deltaW * alpha)
    
    # Append the error for plotting loss curve
    errorList.append(error)
    print(f"Epoch {epoch+1} | Loss {error:-4f}")
```

```python
y_pred = x * w
print(y_pred)
```

```python
# plotting the loss curve
plt.figure(figsize=(8,6))
plt.plot(range(epochs),errorList)
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.xticks(ticks = range(epochs))
plt.title("Epochs vs Loss")
plt.show()
```
